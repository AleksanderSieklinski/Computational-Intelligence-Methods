Proszƒô, korzystajƒÖc z algorytmu roju czƒÖstek znale≈∫ƒá minimum funkcji

$f(x,y)= ùë•^{2} + ùë¶^{2} ‚àí 20(cos ùúãùë• + cos ùúãùë¶ ‚àí 2)$ w przedziale ùë•, ùë¶ ‚àà [‚àí10,10],  


przy za≈Ço≈ºeniu ≈ºe rozwiƒÖzanie jest reprezentowane przez wektor [ùë•ùëñ, ùë¶ùëñ].

Proszƒô zbadaƒá:
- funkcjonowanie algorytmu dla c1=0, c2=2
- funkcjonowanie algorytmu dla c1=2, c2=0 oraz kilku przypadk√≥w po≈õrednich.
- funkcjonowanie algorytmu dla c1 = c2 = 2,2
- funkcjonowanie dla r√≥≈ºnych warto≈õci w.


Za ka≈ºdym razem nale≈ºy podaƒá ≈õrednie wyniki (warto≈õƒá funkcji przystosowania) oraz odchylenie standardowe dla 10 wywo≈Ça≈Ñ algorytmu i przedstawiƒá przyk≈Çadowe przebiegi algorytmu na wykresach (dla jednego z wywo≈Ça≈Ñ).

Nale≈ºy opisaƒá, kt√≥ry wariant sprawdza≈Ç siƒô najlepiej.

```python
import numpy as np
import pyswarms as ps
from pyswarms.utils.functions import single_obj as fx
import matplotlib.pyplot as plt

def f(x):
    return x[:, 0]**2 + x[:, 1]**2 - 20 * (np.cos(np.pi * x[:, 0]) + np.cos(np.pi * x[:, 1]) - 2)
bounds = np.array([[-10, -10], [10, 10]])
params = [
    {'c1': 0, 'c2': 2, 'w': 0.5},
    {'c1': 2, 'c2': 0, 'w': 0.5},
    {'c1': 1, 'c2': 1, 'w': 0.5},
    {'c1': 2, 'c2': 2, 'w': 0.5},
    {'c1': 2, 'c2': 2, 'w': 0.4},
    {'c1': 2, 'c2': 2, 'w': 0.6},
    {'c1': 2, 'c2': 2, 'w': 0.7},
    {'c1': 2, 'c2': 2, 'w': 0.8},
    {'c1': 2, 'c2': 2, 'w': 0.9},
    {'c1': 2, 'c2': 2, 'w': 1.0},
]
for param in params:
    print(f"Testowanie parametr√≥w: {param}")
    results = []
    for _ in range(10):
        optimizer = ps.single.GlobalBestPSO(n_particles=30, dimensions=2, options=param, bounds=bounds)
        cost, pos = optimizer.optimize(f, iters=30)
        results.append(cost)
    print(f"≈örednia warto≈õƒá funkcji celu: {np.mean(results)}")
    print(f"Odchylenie standardowe warto≈õci funkcji celu: {np.std(results)}")
    plt.figure(figsize=(10, 5))
    plt.plot(optimizer.cost_history)
    plt.title(f'Parametry: {param}')
    plt.xlabel('Iteracja')
    plt.ylabel('Warto≈õƒá funkcji celu')
    plt.show()
```

**Wnioski:**
Dla c1=0 i c2=2, algorytm skupia siƒô na ulepszaniu obecnych rozwiƒÖza≈Ñ, ignorujƒÖc eksploracjƒô.
Dla c1=2 i c2=0, algorytm skupia siƒô na eksploracji, ignorujƒÖc ulepszanie obecnych rozwiƒÖza≈Ñ.
Warto≈õƒá w kontroluje prƒôdko≈õƒá czƒÖstek, wiƒôc wp≈Çywa na balans miƒôdzy eksploracjƒÖ a eksploatacjƒÖ.
Algorytm najlepiej sprawdzi≈Ç siƒô dla c1 i c2 = 2 i w = 0.4.

2.

Uzyskany najlepszy wynik proszƒô w miarƒô mo≈ºliwo≈õci por√≥wnaƒá z algorytmem genetycznym ‚Äì dedykowanym dla optymalizacji tej samej funkcji. Zak≈Çadamy u≈ºycie takiej samej liczby epok dla obu algorytm√≥w.
Nale≈ºy por√≥wnaƒá czas dzia≈Çania obydwu algorytm√≥w.

```python
import time
import numpy as np
from geneticalgorithm import geneticalgorithm as ga
import pyswarms as ps

def f(X):
    x = X.reshape(-1, 2)
    return x[:, 0]**2 + x[:, 1]**2 - 20 * (np.cos(np.pi * x[:, 0]) + np.cos(np.pi * x[:, 1]) - 2)
bounds = np.array([[-10, -10], [10, 10]])
start_time = time.time()
options = {'c1': 2, 'c2': 2, 'w': 0.4}
optimizer = ps.single.GlobalBestPSO(n_particles=30, dimensions=2, options=options, bounds=bounds)
cost, pos = optimizer.optimize(f, iters=30)
pso_time = time.time() - start_time
print (f"\nKoszt PSO: {cost}")
print (f"Pozycja PSO: {pos}")
print (f"Czas PSO: {pso_time}")
# Measure the time taken by the genetic algorithm
start_time = time.time()
algorithm_parameters = {'max_num_iteration': 30, 'population_size': 30, 'parents_portion' : 0.3,'mutation_probability' : 0.1, 'crossover_probability' : 0.5, 'elit_ratio' : 0.01, 'crossover_type':'uniform', 'max_iteration_without_improv':None}
model = ga(function=f, dimension=2, variable_type='real', variable_boundaries=np.array([[-10,10],[-10,10]]), algorithm_parameters=algorithm_parameters)
model.run()
ga_time = time.time() - start_time
print(f"\nKoszt GA: {model.output_dict['function']}")
print(f"Pozycja GA: {model.output_dict['variable']}")
print(f"Czas GA: {ga_time}")
```

**Wnioski:**
Czas dzia≈Çania algorytmu PSO jest kr√≥tszy ni≈º algorytmu genetycznego.
PSO mo≈ºe byƒá lepsze dla problem√≥w, kt√≥re wymagajƒÖ szybkiego zbiegania, podczas gdy GA mo≈ºe byƒá lepsze dla problem√≥w, kt√≥re wymagajƒÖ intensywnej eksploracji przestrzeni rozwiƒÖza≈Ñ.
